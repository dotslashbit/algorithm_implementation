```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python

def loss_function(m, b, data):
    total_error = 0
    for i in range(len(data)):
        x = data.iloc[i,0]
        y = data.iloc[i,1]
        total_error += (y - (m * x + b)) ** 2
    total_error / float(len(data))
```

```python
def gradient_descent(m_now, b_now, data, learning_rate):
    m_gradient = 0
    b_gradient = 0
    n = len(data)
    for i in range(n):
        x = data.iloc[i,0]
        y = data.iloc[i,1]
        
        m_gradient += (-2/n) * x * (y - (m_now * x + b_now))
        b_gradient += (-2/n) * (y - (m_now * x + b_now))
        
    m = m_now - m_gradient * learning_rate
    b = b_now - b_gradient * learning_rate
    return m,b

    
```

```python
m = 0
b = 0
l = 0.0001
epochs = 300

for i in range(epochs):
    if i % 50 == 0:
        print(f"Epoch: {i}")
    m, b = gradient_descent(m ,b, data, l)

print(m,b)

plt.scatter(data.iloc[:,0], data.iloc[:,1], color='black')
plt.plot(list(range(20, 80)), [m * x + b for x in range(20, 80)], color='red')
plt.show()
```
